```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
%matplotlib inline
```

# 读取原照片并根据label分割出字段，并将对应的label保存在wordcaption.txt里

```python
datanum=1786
it=1
f1 = open('wordcaption.txt','w')
for i in tqdm(range(datanum)):
    img=cv2.imread('train/'+str(i)+'.jpg')
    f=open('train/'+str(i)+'.txt','r')
    temp=f.read()
    temp=temp.split('\n')
    n=len(temp)
    for j in range(n-1):
        c=temp[j].split(',')
        objects=img[int(float(c[1])):int(float(c[3])),int(float(c[0])):int(float(c[2])),:]
        #objects=cv2.resize(objects, (100, 32))
        cv2.imwrite('image/'+str(it)+'.jpg',objects)
        f1.write(str(it)+' '+c[4]+'\n')
        it+=1
    f.close()
f1.close()
```
